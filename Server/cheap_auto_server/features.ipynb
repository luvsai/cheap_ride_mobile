```python
import numpy as np

def generate_points(top_left, top_right, bottom_left, bottom_right):
    x_min = min(top_left[0], top_right[0], bottom_left[0], bottom_right[0])
    x_max = max(top_left[0], top_right[0], bottom_left[0], bottom_right[0])
    y_min = min(top_left[1], top_right[1], bottom_left[1], bottom_right[1])
    y_max = max(top_left[1], top_right[1], bottom_left[1], bottom_right[1])

    x_coords = np.linspace(x_min, x_max, num=11)
    y_coords = np.linspace(y_min, y_max, num=11)

    points = np.array(np.meshgrid(x_coords, y_coords)).T.reshape(-1,2)
    return points
```

```python
import numpy as np

def divide_world_map(rows, cols):
    lat_step = 180 / rows
    lon_step = 360 / cols

    grid = np.zeros((rows, cols, 2))

    for i in range(rows):
        for j in range(cols):
            grid[i, j, 0] = -90 + (i * lat_step)
            grid[i, j, 1] = -180 + (j * lon_step)

    return grid

```

```python
import numpy as np

def nearest_coordinates(grid, A):
    # Compute the Euclidean distance between each coordinate in the grid and A
    distances = np.linalg.norm(grid - A, axis=-1)
    # Get the indices of the 4 smallest distances (i.e. the 4 nearest coordinates)
    nearest_indices = np.argpartition(distances.flatten(), 4)[:4]
    # Get the corresponding coordinates from the grid
    nearest_coordinates = grid[np.unravel_index(nearest_indices, distances.shape)]
    return nearest_coordinates

```
